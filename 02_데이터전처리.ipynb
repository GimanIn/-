```python
# 정규화

# 표준화(Standardization)

import seaborn as sns

iris = sns.load_dataset('iris')
iris.iloc[:, :-1]
x = iris.iloc[:, :-1]
x
```

```python
y = iris.iloc[:, -1]
y
```

```python
# 데이터 전처리 : scaling

from sklearn.preprocessing import scale

scale(x)
```

```python
x_scale = scale(x)
```

```python
x_scale[:5, :]
```

```python
x_scale.mean(axis=0)
# axis = 0 (서로 다른 행끼리)
# axis = 1 (서로 다른 열끼)
# default(기본값) : axis=0
```

```python
# minmax_scale()

from sklearn.preprocessing import minmax_scale

sc = minmax_scale(x)
sc[:5, :]
# 0 ~ 1 사이 값으로 스케일링
```

```python
x[:5]
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

sc.fit(x) # 표준화(평균, 표준편차 계산)
```

```python
iris_scale = sc.transform(x)
iris_scale[:5, :]
```

```python
# 스케일 백 (scale back) 원래 값 알고 싶어

sc.inverse_transform(iris_scale)[:5, :]
```

```python
titanic.groupby('sex')['survived'].mean()
```

```python
sex_mean = titanic.groupby('sex')['survived'].mean()

titanic['sex'].map(sex_mean)
```

```python
titanic['sex_mean'] = titanic['sex'].map(sex_mean)

titanic[:5]
```

```python
# sex, sex_mean 변수 추출

titanic[['sex','sex_mean']].head()
```

```python
# pivot table 정리

titanic_pivot =\
titanic.pivot_table(index = titanic.index,
                    columns= "survived",
                    values='sex_mean')
```

```python
titanic_pivot[:5]
```

```python
titanic_pivot.hist(bins=10)
```

```python
# nan 생성

import random

random.seed(42)

for col in range(4):
  x.iloc[[random.sample(range(len(iris)), 20)],col] = float('nan')
```

```python
x.head()
```

```python
x.mean(axis=0)
```

```python
# imputer 활용, nan(결측치) 해결(평균, 중위수로 채움)

from sklearn.impute import SimpleImputer

imp_mean = SimpleImputer(strategy='mean')
imp_mean.fit(x)
```

```python
imp_mean.transform(x)[:5, :]
```

```python
# 중위수(중앙값) 대체

x.head()
```

```python
x.median(axis=0)
```

```python
imp_median = SimpleImputer(strategy='median')
```

```python
imp_median.fit(x)
```

```python
imp_median.transform(x)[:5, :]
```

```python
# mode(최빈값) 대체

x.mode(axis=0)
```

```python
imp_freq = SimpleImputer(strategy='most_frequent')
imp_mostfreq = imp_freq.fit_transform(x)
imp_mostfreq[:5, :]
```
